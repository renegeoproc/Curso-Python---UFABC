# 1. Justificativa

A presente atividade visa desenvolver uma análise de dados  sobre as mudanças climáticas e desastres naturais no Brasil através das informações disponíveis no Atlas digital de Desastres no Brasil a partir da manipulação e análise dos dados geoespaciais utilizando Python.

O ordenamento territorial no Brasil ainda é conduzido, em grande parte, através de leis urbanísticas ao invés de planos urbanísticos. A falta de articulação entre instrumentos de planejamento e controle urbano abrem brechas para a consolidação de expansão uma urbana sem planejamento, com aumento da ocupação de áreas ocupadas suscetíveis à ocorrência de desastres climáticos, estes cada vez mais frequentes e extremos em decorrência das mudanças climáticas, vitimando centenas de vidas e resultando em elevados danos materiais e socioambientais.

Muito embora a força desses eventos climáticos seja a maior responsável por esse cenário catastrófico, não se pode negar que a ausência de ferramentas que assegurem o direito à vida e à moradia digna, conforme estabelece a Constituição Federal do Brasil, confere ao poder público a sua parcela de culpa. De acordo com Huitema et al. (2018), as cidades representam um ponto chave nas discussões sobre o enfrentamento das mudanças climáticas, atuando principalmente na adaptação climática, que se mostra a alternativa mais viável de enfrentamento considerando os efeitos de curto prazo que beneficiam populações e o meio ambiente a nível local, atingidos diretamente pelas mudanças climáticas.

A questão da gestão do desastre no País está atrelada aos marcos legais da  formulação da Política Nacional de Proteção e Defesa Civil e do Estatuto da Cidade. A primeira foi instituída por meio da Lei 12.608, de abril de 2012 e reformada na pela Lei 14.750, de 12 de dezembro de 2023, que foi criada com o objetivo de dar uma resposta à sociedade em relação às milhares de mortes e prejuízos decorrentes de desastres. O Estatuto das Cidades, Lei 10.257 de 10 de julho de 2001 previa em seu texto que municípios com mais de 20.000 habitantes deveriam ter planos diretores, que deveriam ser a lei de referência de  planejamento do território dos municípios e que nas novas áreas de expansão urbana deveriam ter estudos sobre a susceptividade ao risco e ter capacidade de suporte da infraestrutura.  Não é por falta de lei, pois mesmo com a evolução legal, com a previsão da responsabilização por riscos decorrentes da ação humana, a gestão do risco no país ainda está muito longe da realidade  das áreas vulneráveis.
No Brasil, a gestão de risco enfrenta desafios significativos decorrentes da complexidade e diversidade de suas dimensões. Um dos principais problemas reside na falta de uma legislação específica que regulamente de forma abrangente e eficaz a gestão de riscos em diferentes setores da sociedade. Isso cria lacunas na prevenção e resposta a eventos adversos, deixando o país suscetível a crises e impactos socioeconômicos devastadores.
Segundo Barros (2017), a capacidade de adaptação das cidades envolve a integração entre entre setores municipais como clima, gestão de recursos hídricos, mobilidade urbana, e gestão de desastres. O planejamento urbano também figura como um setor fundamental diretamente associado à adaptação climática, considerando que este é responsável por promover a regulação da ocupação e uso do solo de áreas urbanas, identificando e viabilizando a redução de vulnerabilidades e riscos associados às mudanças climáticas (Wamsler et al., 2013).

O Sexto Relatório de Avaliação (AR6), lançado em março de 2023 pelo Painel Intergovernamental sobre Mudanças Climáticas (IPCC) concluiu que a temperatura média global  deveria atingir ou ultrapassar 1,5ºC nas próximas duas décadas. O relatório Global Climate Highlights, lançado pela Agência Espacial Europeia em janeiro deste ano, confirmou que em 2023 atingimos a marca de 1,48ºC em relação aos níveis pré-industriais, muito próximo do previsto.

Milhões de pessoas já enfrentam as consequências reais decorrentes do aumento das temperaturas. Um relatório publicado pelo Centro de Ciência e Política de Mudança Climática Gerard J. Mangone, da Universidade de Delaware, nos Estados Unidos, aponta que em 2022, o valor estimado em perdas econômicas resultante das mudanças climáticas foi de US$1,5 trilhões, sendo países com Produto Interno Bruto (PIB) de baixa renda, dependentes de agricultura e as regiões tropicais as mais afetadas.

A Organização das Nações Unidas (ONU) utiliza o termo “perdas e danos” para referir-se às consequências das mudanças climáticas. Muito embora exista uma definição precisa, os “danos” podem ser entendidos como aqueles que podem alterar a paisagem de forma permanente, como a desertificação de territórios ou o avanço do mar sobre pequenas ilhas, por exemplo, ao passo que perdas podem ser classificadas em duas principais categorias: econômicas, afetando recursos, bens e serviços comercializados, e não econômicas, como o desaparecimento de familiares, culturas, ou modos de vida (WRI, 2022).

```python
#Plotando dados em gráfico de linhas
fig, ax = plt.subplots(2, 1, figsize=(10,10))

ax[0].plot(cotas_inundacao['Ano'], cotas_inundacao['Cota Máxima Atingida'], label='Cota de Inundação Máxima')
ax[0].set_title('Nível Máximo Atingido (em metros)')
ax[0].set_ylabel('Nível do Rio Acre em Brasiléia')

ax[1].plot(cotas_inundacao['Ano'], cotas_inundacao['Pessoas afetadas'], label= 'Pessoas Afetadas')
ax[1].set_title('Pessoas atingidas pelas cheias e inundações')
ax[1].set_ylabel('Pessoas atingidas')
```

```python
#Cruzando dados de Danos Materiais e Danos Humanos
perdas_danos['data'] = pd.to_datetime(perdas_danos['data'])
perdas_danos['ano'] = perdas_danos['data'].dt.year
perdas_danos_anos = perdas_danos[perdas_danos['ano'].isin([2012, 2015])]

fig, ax1 = plt.subplots()


ax1.plot(perdas_danos_anos['ano'], perdas_danos_anos['total_danos_materiais'], color='b', marker='o')
ax1.set_xlabel('Ano')
ax1.set_ylabel('Total de Danos Materiais (R$)', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()

ax2.bar(perdas_danos_anos['ano'], perdas_danos_anos['total_danos_humanos'], color='orange', alpha=0.7)
ax2.set_ylabel('Total de Danos Humanos', color='orange')
ax2.tick_params('y', colors='orange')

formatter = ticker.FuncFormatter(lambda x, pos: 'R$ {:,.2f}'.format(x))
ax1.yaxis.set_major_formatter(formatter)

plt.title('Total de Danos Materiais e Humanos em 2012 e 2015')

plt.show()
```

# 5. Conclusão ou considerações finais

A análise dos efeitos das enchentes e inundações, juntamente com a ausência de instrumentos eficazes de gestão de risco, revela uma situação complexa para o município de Brasiléia. Ao longo dos anos, eventos climáticos extremos e recorrentes, em sua maioria de natureza hidrológica, causam danos significativos à cidade, impactando não apenas a infraestrutura urbana, mas também a população local e a economia regional. Além dos danos às unidades habitacionais, o município - e o Estado - se vêem isolados do resto do país em decorrência de danos causados às estradas responsáveis pela chegada e escoamento de produtos essenciais necessários.

Os registros indicam que Brasiléia havia enfrentado as piores enchentes em 2012 e 2015, até então. Contudo, em março desse ano (2024), os estragos já superam os dos anos anteriores. Em 2012, por exemplo, a área urbana do município foi severamente afetada, com aproximadamente 95% do território urbano endo inundado. O trecho do Rio Acre que corta o município alcançou a cota de 13,60 metros.

A situação se agravou em 2015, quando o nível do Rio Acre ultrapassou a enchente histórica de 2012, atingindo a cota de 15,55 metros. Em março deste ano (2024), o rio chegou a 15,58 metros. Esses episódios ressaltam a recorrência dos eventos extremos e a incapacidade da cidade em lidar eficazmente com essas situações. A falta de instrumentos adequados de gestão de risco agrava ainda mais a situação. O Plano Diretor do município, aprovado em 2013, não oferece medidas concretas de mitigação ou enfrentamento das enchentes sazonais que assolam a região. Além disso, o plano não apresenta qualquer mapeamento ou zoneamento na cidade, estudo técnico ou mesmo prévio que embase as decisões nele apresentadas. Um integrante do grupo tentou contato com a prefeitura e orgãos de planejamento responsáveis, porém não obtever nenhum retorno sobre a disponibilidade - ou existência - desses dados.  Essa lacuna crítica no preparo técnico para o planejamento e enfrentamento de eventos climáticos extremos são evidenciados nos dados.

É importante pontuar que mesmo os dados divulgados no Atlas Digital de Desastres no Brasil possui lacunas que são difícies de serem preenchidas mesmo com técnicas de machine learning. Estimar valor em danos materiais ou mesmo humanos sem ter acesso às ferramentas de gestão e risco em vigência ou às ações já tomadas por parte do poder público para atenuar os impactos desses eventos se torna uma tarefa praticamente impossível.

Embora os dados fornecidos ofereçam uma visão parcial dos impactos das enchentes e inundações, é importante ressaltar que eles não capturam completamente a extensão dos danos causados. Isso se torna evidente ao analisarmos o número de pessoas afetadas pelos eventos. Em 2012, houve o registro de 3486 pessoas desabrigadas, que perderam suas residências e foram realojadas em abrigos públicos, além de 3367 pessoas desalojadas, que foram obrigadas a deixar suas casas, mas não necessariamente as perderam. Em contrapartida, em 2015, apesar da contraditória diminuição nos números, considerando que o nível alcançado pelo rio foi maior que em 2012 - com 804 pessoas desabrigadas e 1232 desalojadas -, é importante notar que outras 8.304 pessoas foram consideradas como afetadas diretamente pela enchente, embora não tenha sido especificado de que forma foram impactadas. Além disso, quantidade de unidades habitacionais danificadas e o valor em danos às unidades habitaiconais em 2015 é significativamente maior que em 2012. Em 2015, foram 2072 unidades habitacionais danificadas, somando mais de 18 milhões de reais em prejuízos, enquanto que em 2012, foram 184 unidades habitacionais danificadas (4 milhões de reais em prejuízos).

Embora exista um Plano de Gestão e Risco estadual, não há evidências claras de implementação efetiva dessas medidas em Brasiléia, e a falta de coordenação e implementação de políticas de gestão de risco em nível municipal contribui para a vulnerabilidade da cidade a eventos extremos, além de resultar em constantes perdas de atividades econômicas locais, danos à infraestrutura e retardamento no desenvolvimento da região afetadas, impactando diretamente o Produto Interno Bruto (PIB) do município e do Estado.

Quando olhamos para as áreas de risco setorizadas pelo Serviço Geológico do Brasil (SGB), podemos notar que algumas áreas permanecem classificadas como altamente suscetíveis aos impactos dos eventos climáticos extremos durante os anos e, mesmo assim, seguem ocupadas pela população. Essa persistência na ocupação dessas áreas de risco demonstra uma deficiência na gestão do poder público em relação à proteção dos cidadãos e ao ordenamento territorial.

Segundo o CPRM (2015), o centro histórico do município de Brasiléia ocupa uma antiga planície aluvionar, altamente suscetível a eventos de cheias pronunciadas. Outro agravante é a característica geológica do município, constituído por solo argiloso e de baixa permeabilidade, o que dificulta a infiltração das águas pluviais, resultando no escoamento direto na bacia de drenagem, elevando rapidamente o nível fluvial dos rio e igarapés. Problemas de investimento em infraestrutura resiliente em áreas suscetíveis à eventos climáticos extremos contribui para a perpetuação dos riscos e agravamento dos impactos. A ausência de sistemas de alerta precoce, abrigos seguros e rotas de evacuação eficientes aumenta a exposição da população aos perigos das enchentes e inundações, colocando vidas em risco. A falta de planejamento e gestão integrada do território, a inexistência de políticas coordenadas entre os diferentes órgãos governamentais e níveis de governo resulta em uma abordagem fragmentada e desarticulada para lidar com os desafios de gestão de risco.

Diante desse cenário, é imprescindível a implementação imediata de medidas eficazes de gestão de risco e adaptação para enfrentar os desafios impostos pelos eventos climáticos extremos em Brasiléia. Isso inclui a revisão e atualização do Plano Diretor municipal, a implementação de estratégias de mapeamento de áreas de risco e a coordenação entre os níveis municipal e estadual para garantir a efetivação de políticas públicas que visem a proteção da população e a resiliência da cidade diante dos desastres naturais recorrentes.
